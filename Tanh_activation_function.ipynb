<strong>Sigmoid VS Tanh</strong> which one is better? The answer is Hyperbolic Tangent AKA (Tanh). They both have the properties of good activation functions like non-linear, continuous, differentiable. What makes tanh better is ZERO-CENTERED. In simple terms: A function is said to be zero-centered when its range contains both positive and negative values. (Sorry for not diving deep into zero-centered). Just a small takeaway is, if the function is not zero-centered, to optimize the weight at its best, we will have to invest more time (time complexity will be high). As we already know, the sigmoid's range is between 0 to 1 so it will always return the positive value and it does not have a negative side. (Seems not having a negative side is only good for human beings :p ). But Tanh has a range between -1 to 1. That means clearly Tanh is zero-centered and sigmoid is not.
```python
def tanh_function(z):
    return (np.exp(z)-np.exp(-1*z))/(np.exp(z)+np.exp(-1*z))
```

```python
def gradient_tanh(z):
    return 1- np.square(tanh_function(z))
```

If we look at the output of tanh activation function, it has squeezed the value in between -1 to 1.

```python
z = [x for x in range(-15,16)]
tanh_output = [tanh_function(y) for y in z]
plt.plot(z,tanh_output)
```

```python
z = [x for x in range(-15,16)]
dtanh_output = [gradient_tanh(y) for y in z]
plt.plot(z,dtanh_output)
```
